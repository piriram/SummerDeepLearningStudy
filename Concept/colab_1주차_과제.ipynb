### ÌçºÏÖâÌä∏Î°†

#### Îã®Ïùº ÌçºÏÖâÌä∏Î°† üî•

``` python
Í∞ÑÎã®Ìïú Îã®Ïùº ÌçºÏÖâÌä∏Î°† ÌÅ¥ÎûòÏä§Î•º Íµ¨ÌòÑÌï¥Î¥ÖÏãúÎã§.

ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Ïóê Í∞ÄÏ§ëÏπò(Weight)Î•º ÏÑ†ÌòïÍ≥± Ìï¥Ï£ºÍ≥†, Ìé∏Ìñ•(bias)ÏùÑ ÎçîÌï¥Ï£ºÎäî forward Ìï®ÏàòÎ•º ÏôÑÏÑ±ÌïòÎ©¥ Îê©ÎãàÎã§.
Ïù¥Îïå, thesholdÎ•º ÎÑòÏúºÎ©¥ 1 ÎÑòÏßÄ ÏïäÏúºÎ©¥ 0ÏùÑ Î∞òÌôòÌïòÎäî step functionÍπåÏßÄ Ìè¨Ìï®Ìï¥Ïïº Ìï©ÎãàÎã§.
```


````python
class Perceptron:
    def __init__(self, weight, bias):  # __init__ : Ï¥àÍ∏∞ÌôîÎ•º ÏúÑÌïú Ìï®Ïàò, ÌÅ¥ÎûòÏä§Î•º ÏÉùÏÑ±ÌïòÍ≥† ÌôúÏö©ÌïòÎ†§Î©¥ Íº≠ ÏûëÏÑ±Ìï¥Ï§òÏïº Ìï©ÎãàÎã§.

        self.weight = weight  # Í∞ÄÏ§ëÏπò Î≥ÄÏàò Ï†ïÏùò
        self.bias = bias  # Ìé∏Ìñ• Î≥ÄÏàò Ï†ïÏùò

    #########################TODO######################
    def forward(self, x):
        y = self.weight * x + self.bias
    #########################TODO######################
        return y
````

````python
# ÏõêÌïòÎäî Í∞ÄÏ§ëÏπò, Ìé∏Ìñ•, ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º Ï†ïÏùòÌï¥Ï£ºÏÑ∏Ïöî!
weight = 2
bias = 3
input = 3


# ÏïÑÎûò ÏΩîÎìúÎäî ÏàòÏ†ïÌïòÏã§ ÌïÑÏöîÍ∞Ä ÏóÜÏäµÎãàÎã§!
p1 = Perceptron(weight, bias)
````

##### Ï†ïÎãµ ÌôïÏù∏

````python
if p1.forward(input) == input * weight + bias:
    print("üéâüéâüéâ ÏÑ±Í≥µÏûÖÎãàÎã§!!! üéâüéâüéâ")
else:
    print("üî•Îã§Ïãú ÎèÑÏ†ÑÌï¥Î¥ÖÏãúÎã§!üî•")
````

#### Îã§Ï∏µ ÌçºÏÖâÌä∏Î°† üî•

``` python
Ïù¥Î≤àÏóî Îã§Ï∏µ ÌçºÏÖâÌä∏Î°†(MLP, Multi-Layer Perceptron) ÌÅ¥ÎûòÏä§Î•º Íµ¨ÌòÑÌï¥Î¥ÖÏãúÎã§.
Íµ¨ÌòÑÌï† MLPÏùò Íµ¨Ï°∞Îäî ÏïÑÎûò Í∑∏Î¶ºÍ≥º Í∞ôÏäµÎãàÎã§.

Îã®Ïùº ÌçºÏÖâÌä∏Î°†Í≥º ÎßàÏ∞¨Í∞ÄÏßÄÎ°ú ÌÅ¥ÎûòÏä§Î•º ÏôÑÏÑ±ÌïòÎ©¥ Îê©ÎãàÎã§.

```
image.png


````python
class MLP:
    def __init__(self, input_dim, hidden_dim, output_dim):
        self.weight1 = np.random.rand(input_dim, hidden_dim)  # Ïù∏ÏûêÎ°ú Î∞õÏùÄ ÌÅ¨Í∏∞Ïùò 0~1 ÏÇ¨Ïù¥ Î¨¥ÏûëÏúÑ Í∞íÏùÑ Î∞òÌôòÌï¥Ï§çÎãàÎã§.
        self.weight2 = np.random.rand(hidden_dim, output_dim)

    #########################TODO######################
    def forward(self, input):
        x = np.dot(input,self.weight1)
        x = np.dot(x,self.weight2)

    #########################TODO######################
        return x
````

````python
# Í∞Å Ï∏µÏùò Ï∞®Ïõê Ï†ïÏùò
input_dim = 3
hidden_dim = 3
output_dim = 2

input = np.random.rand(3) # Î¨¥ÏûëÏúÑ ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±

# ÏïÑÎûò ÏΩîÎìúÎäî ÏàòÏ†ïÌïòÏã§ ÌïÑÏöîÍ∞Ä ÏóÜÏäµÎãàÎã§!
mlp  = MLP(input_dim, hidden_dim, output_dim)
result = mlp.forward(input)
print(result)
````

##### Ï†ïÎãµ ÌôïÏù∏

````python
if mlp.forward(input).size == 2:
    print("üéâüéâüéâ ÏÑ±Í≥µÏûÖÎãàÎã§!!! üéâüéâüéâ")
else:
    print("üî•Îã§Ïãú ÎèÑÏ†ÑÌï¥Î¥ÖÏãúÎã§!üî•")
````

``` python
Ï∂úÎ†•Ï∏µÏóêÏÑú Ï£ºÎ°ú ÏÇ¨Ïö©ÎêòÎäî Ìï®ÏàòÎ°ú, Îã§Ï§ë Î∂ÑÎ•òÎ•º ÏúÑÌïú Î™®Îç∏Ïùò ÎßàÏßÄÎßâ LayerÏóê ÌùîÌûà Îì±Ïû•ÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§.
```
